## Introduction: Business Problem <a name="introduction"></a>

In Germany it is exceptionally difficult to find a place in a kindergarden, due to the low numbers of kindergardens and the increase in births.

In this project we will try to find the best neighbourhoods in Hamburg for new parents, in terms of kindergarden availability. Specifically, this report will be targeted to real estate offices in Hamburg.

We are particularly interested in **areas that have high numbers of Preschools**, so that the parents propability to find a kindergarden close to their house will be increased.

We would also prefer locations that include **Playgrounds, Parks, Pools, Restaurants and Metro Stations**, assuming that first condition is met.

Advantages of each area will then be clearly expressed so that best neighbourhoods can be suggested to the clients by the real estate offices.
